Scenario:

-Use len() and string methods (like .isalpha() ,lower(),upper()).

-Replace all spaces with underscores.

-Show only every 3rd character of the message.

-Use slicing: message[::3]

-Reverse the message and display it.

-Display the cleaned and transformed result.

```python
enter_your_secret_msg=input("Pls Enter the Secret message:")
print("My secret msg is:",enter_your_secret_msg)
print("My secret msg length is:",len(enter_your_secret_msg))
upper_txt=enter_your_secret_msg.upper()
lower_txt=enter_your_secret_msg.lower()
print("My secret msg is alpha:",enter_your_secret_msg.isalpha())
print("My Secret msg is in lowercase:",lower_txt)
print("My Secret msg is in uppercase:",upper_txt)
Replace_underscore=upper_txt.replace(" ","")
print(f"scret_msg:{Replace_underscore}")
enter_your_secret_msg[::-3]
s1=slice(3)
print(enter_your_secret_msg[s1])
enter_your_secret_msg[::-1]
print(f"secret_msg:{Replace_underscore}")
```

Project Title: Retail Inventory Manager

***Scenario:***

 You are managing an inventory system for a retail shop. Certain product categories are fixed, but the actual products can be added or removed by the staff.

***Tasks:***

Create a tuple with fixed product categories like Electronics, Stationery, and Clothing.
Create a list of products that can be updated dynamically.

Display the final list of products and the product categories.

```python
# Step 1: Create Tuple for each student (immutable personal info)

student1=("Sidhu", 1, "Viscom")
student2=("Anu", 2, "MBA")
student3=("Shravan", 3, "BSc")

all_students=[student1,student2,student3]
all_students

# Step 2: Create 1D List for marks of 3 student

marks1=[90, 98, 67,87,88]
marks2=[88, 70, 82,78,80]
marks3=[98, 79, 77,67,85]

# Step 3: Combine all marks into a 2D List

first_batch=[marks1,marks2,marks3]

second_batch=[
    [77, 78, 79,80,81],
    [82, 83, 84,85,86],
    [87, 88, 89,90,91]]

# Step 4: Combine all batches (3D List - two different batches)

all_batches=[first_batch,second_batch]

print(all_batches)
```
