# Case Study Scholarship Eligibility Checker

# Instructions:

1. Create variables: student_name, marks, attendance, family_income.

2. Use conditional statements (if, elif, else) to:

    Award Full Scholarship if marks ≥ 90 and attendance ≥ 90.
    Award Half Scholarship if marks ≥ 80 or attendance ≥ 85.
    Award Need-Based Scholarship if marks ≥ 70 and family_income < 30000.
    Otherwise, display Not Eligible.

3. Display all data (no loops).

```python
# Creating variables
student_name = "Kalanithi A"
mark = 80
attendance = 90
family_income = 30000

# Display student details
print("----- Student Details -----")
print("Name:", student_name)
print("Mark:", mark)
print("Attendance:", attendance)
print("Family Income:", family_income)

# Scholarship Eligibility Conditions
print("\n----- Scholarship Result -----")

if mark >= 90 and attendance >= 90:
    print("Award: Full Scholarship")

elif mark >= 80 or attendance >= 85:
    print("Award: Half Scholarship")

elif mark >= 70 and family_income < 30000:
    print("Award: Need-Based Scholarship")

else:
    print("Award: Not Eligible")
```
