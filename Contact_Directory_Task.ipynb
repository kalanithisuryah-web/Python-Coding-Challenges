# List Methods and Slicing

**Project Title: Contact Directory**


Scenario:

 You have been assigned to build a contact management system where employees can store and retrieve contact details efficiently.



Tasks:

Create a 2D list where each row holds a contactâ€™s name, phone number, and email address.

Allow the user to add a new contact.

Display all contacts sorted alphabetically by name.

Allow the user to search for contacts by name. The search should support partial matches.

After sorting, display only the first 3 contacts in the list.


**--------------------------------------------------------------------------------------------------**

```python
contacts=[
    ["Siva",9895624115,"siva@gmail.com"],
    ["Aishu",9895441218,"aishu@gmail.com"],
    ["Sushmitha",9888425166,"sushmitha@gmail.com"]
]

print("Add a New Contact")
name=input("Enter your name: ")
phone_no=int(input("Please Enter your Phone Number: "))
email_address=input("Fillup the Email Address: ")

contacts.append([name, phone_no, email_address])
print("\nContact added successfully!\n")

contacts.sort(key=lambda x: x[0])
print("All Contacts (Sorted Alphabetically):")
for c in contacts:
  print(f"Name: {c[0]}, Phone: {c[1]}, Email: {c[2]}")

search=input("\nEnter name to search: ").lower()

print("\nSearch Results:")
found = False
for c in contacts:
    if search in c[0].lower():
        print(f"Name: {c[0]}, Phone: {c[1]}, Email: {c[2]}")
        found = True

if not found:
    print("No contacts found!")


print("\nFirst 3 contacts:")
first_three=contacts[:3]
for c in first_three:
    print(f"Name: {c[0]}, Phone: {c[1]}, Email: {c[2]}")


```
